```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[10,20,25,30])
plt.xlim(0,5)
plt.ylim(0,35)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('sample plot')
plt.show()
```
